```python
from bs4 import BeautifulSoup
import pandas as pd
import requests


# You can add here website link which are showing you stocks historical data in tabular format. 
url = "https://www.indiainfoline.com/company/yes-bank-ltd-share-price/share-price-historical-data/25267"
# url = 'https://in.investing.com/equities/yes-bank-historical-data?end_date=1686422196&interval_sec=monthly&st_date=1457634600'
html_data = requests.get(url).text

soup = BeautifulSoup(html_data, 'html5lib')
```

```python
# Now we will be extracting table headers for the dataframe columns

''' If there are multiple tables on the website, you can find all 
    table and iterate through them to find the one you want to 
    extract using a unique keyword around the table.

    *Note: You have to inspect the or check the list of tables 
    one by one and pass the table_index for further extraction ,
    if you're unable to find the exact table using keyword.

    You can visit 'https://beautiful-soup-4.readthedocs.io/en/latest/'
    for more information on fuctions to parse the HTML doc 
'''

unique_keyword = 'Total Turnover'
# unique_keyword = ''
tables = soup.find_all('table')
table_index = 0
for i, table in enumerate(tables):
    if unique_keyword in table:
        table_index = i

# print(tables[table_index])

headers = []

for col in tables[table_index].find('thead').find('tr'):
    headers.append(col.text.strip())
headers = [x for x in headers if x != '']
data_df = pd.DataFrame(columns=headers)
data_df.columns


for row in tables[table_index].find('tbody').find_all('tr'):
    
    col = row.find_all('td')

    new_row = []

    for i in range(len(headers)):
        new_row.append(col[i].text.strip())
    
    data_df = pd.concat([data_df, pd.DataFrame([new_row], columns=headers)], ignore_index=True)
```
